```python
def Weight_group(Weight):
    
    
    Weight = int(Weight)
    
    if Weight < 10:
        bucket = '<10'
    
    if Weight in range(10, 20):
        bucket = '10-20'
        
    if Weight in range(20, 30):
        bucket = '20-30'
        
    if Weight in range(30, 40):
        bucket = '30-40'
        
    if Weight in range(40, 50):
        bucket = '40-50'
        
    if Weight in range(50, 60):
        bucket = '50-60'
   
    if Weight >= 60:
        bucket = '60+'

    return bucket
```

```python
df1['WEIGHT_W23']=df1['WEIGHT_W23'].apply(Weight_group)

df1["WEIGHT_W23"].head(3)
```

```python
import statsmodels.api as sm # import statsmodels 

X = df["F_RACETHN_RECRUITMENT"] ## X usually means our input variables (or independent variables)
y = df["THERMO2_THERMTRUMP_W23"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()
```
